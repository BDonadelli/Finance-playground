```python
acoes = 'PETR4.SA ABEV3.SA VALE3.SA'
```

```python
tickers = yf.Tickers(acoes)
```

```python
tickers.tickers["PETR4.SA"].balance_sheet
```

```python
tickers = yf.Tickers(acoes.split())
tickers.tickers
```

```python
dict_data = {}

for ticker in tickers.tickers.keys():
    print("Baixando: {}".format(ticker))
    dict_data[ticker] = tickers.tickers[ticker].history(period="max")
```

```python
dict_data["PETR4.SA"]
```

```python
dict_data[ticker]["Close"]
```

```python
df_data = pd.DataFrame()

for ticker in dict_data.keys():
    df_aux = dict_data[ticker].rename(columns={"Close": ticker})
    df_data = df_data.join(df_aux[ticker], how="outer")

df_data   
```

```python
df_data
```

```python
df_ret = df_data / df_data.shift(1) - 1
```

#####  pior retorno

```python
df_ret.idxmin()
```

```python
df_ret.min()
```

```python
df_ret
```

#### retornos do dia seguinte a uma queda

```python
df_ret[df_ret.shift(1) < 0].mean()
```

```python
df_ret[df_ret.shift(1) < 0].std()
```

```python
df_study = df_ret[df_ret.shift(1) < 0]

print("MÃ©dia e desvio padrÃ£o dos retornos apÃ³s dias de queda")
for acao in acoes.split():
    print(f'{acao} - {df_study[acao].mean()*100} - {df_study[acao].std()*10}')

fig, ax = plt.subplots(figsize=(20, 8))
sns.histplot(df_study, kde=True)

```

#### se as quedas foram  maiores que 5%.

```python
df_study = df_ret[df_ret.shift(1) < -0.05]


print("MÃ©dia e desvio padrÃ£o dos retornos apÃ³s dias de queda")
for acao in acoes.split():
    print(f'{acao} - {df_study[acao].mean()*100} - {df_study[acao].std()*10}')



fig, ax = plt.subplots(figsize=(20, 8))
sns.histplot(df_study, kde=True)
```
