```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Define o código Anbima do fundo
cod_fundo = "C0000463698"
url = f"https://data.anbima.com.br/fundos/{cod_fundo}/dados-periodicos"

# Inicializa o navegador
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Executa sem abrir janela
options.add_argument("--disable-gpu")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
```

```python
try:
    print("Carregando página...", url)
    driver.get(url)

    # Dá um tempo para a página carregar os dados dinâmicos
    time.sleep(5)

    # Localiza a tabela de dados periódicos
    table = driver.find_element(By.TAG_NAME, "table")
    rows = table.find_elements(By.TAG_NAME, "tr")

    # Extrai as colunas de interesse da tabela
    data = []
    for row in rows[1:]:  # pulando o header
        cols = row.find_elements(By.TAG_NAME, "td")
        if len(cols) >= 2:
            data_ref = cols[0].text.strip()
            valor_cota = cols[2].text.strip()
            data.append((data_ref, valor_cota))

    # Converte para um DataFrame
    df = pd.DataFrame(data, columns=["data", "valor_cota"])
    print("\nDados extraídos:\n", df)

except Exception as e:
    print("Erro ao extrair os dados:", e)

finally:
    driver.quit()
```
