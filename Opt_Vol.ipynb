```python
import os
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

data_path = str(os.getcwd()) + r"/data/"

opts = webdriver.ChromeOptions()
opts.add_experimental_option("detach", True)
opts.add_experimental_option("prefs", {
  "download.default_directory": data_path,
  "download.prompt_for_download": False,
  "download.directory_upgrade": True,
  "safebrowsing.enabled": True
})
driver=webdriver.Chrome(options=opts)

driver.get("https://opcoes.net.br/acoes")
```

```python
login_botao = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.LINK_TEXT, "Efetuar login"))
    )
login_botao.click()

driver.find_element(By.ID, "CPF").send_keys("115.723.148-93")
driver.find_element(By.ID, "Password").send_keys("opcoes04658")

# Clicar no botão de login
botao_login = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, '//button[@type="submit"]'))
)
botao_login.click()

WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.LINK_TEXT, "Continuar com dados de fechamento"))
).click()

from selenium.webdriver.support.ui import Select

# Esperar o elemento <select> com ID "defaultFilter" estar presente
select_element = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID, "defaultFilter"))
)

# Criar objeto Select
select_obj = Select(select_element)

# Selecionar pelo texto visível
select_obj.select_by_visible_text("Ativos com opções")
```
