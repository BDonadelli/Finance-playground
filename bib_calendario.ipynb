```python
from bizdays import Calendar
from datetime import datetime
import pandas as pd
```

```python
def calendario_anbima():
    # Extrai a lista de feriados direto do site da anbima
    holidays = pd.read_excel('http://www.anbima.com.br/feriados/arqs/feriados_nacionais.xls', 
                             skipfooter=9)["Data"]
    weekends = ["Saturday", "Sunday"]
    # Cria o calend√°rio
    calendar = Calendar(holidays, weekends)
    return calendar
```

```python
cal = calendario_anbima()
```

```python
cal.bizdays(datetime(2023, 1, 1), datetime(2023, 12, 31))
```
